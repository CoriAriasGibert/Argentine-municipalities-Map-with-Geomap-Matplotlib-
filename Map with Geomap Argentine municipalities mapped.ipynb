```python
# Import geopandas library
import geopandas as gpd

# Load file
url = "https://infra.datos.gob.ar/catalog/modernizacion/dataset/7/distribution/7.14/download/municipios.geojson"
geoData = gpd.read_file(url)
```

```python
geoData.head(2)
```

```python
# Import the geoplot library.
import matplotlib.pyplot as plt

# Set the plot size for this notebook:
plt.rcParams["figure.figsize"]=13,13

# Draw a map with matplotlib
geoData.plot(color="pink", edgecolor='deeppink', linewidth=.5);
plt.axis('on');
```
